# Data exploration

The purpose of this notebook is to examine a dataset. The data selected is from SCAG's public database. 

```python
# importing necessary packages
# pandas is the only one needed for the purpose of examining this data
import pandas as pd
import geopandas as gdp
```

```python
# input: loading the data/csv file
# output: nothing visible happens here but the data should be loaded
data = pd.read_csv('Bike_Routes_SCAG_Region.csv')
```

```python
#showing the first 5 rows of the data
data.head()
```

```python
#getting a count for unique values in column
class_E=data['CLASS_E'].value_counts()
class_E
```

```python
#getting a count for unique values in column
class_P=data['CLASS_P'].value_counts()
class_P
```

```python
#checking the type of data the value_counts() generates
type(class_P)
```

```python
#changing the data from a series to a dataframe
class_E = class_E.reset_index()
class_E
```

```python
#changing the data from a series to a dataframe
class_P = class_P.reset_index()
class_P
```

```python
#renaming the columns
class_E.columns=['class_type', 'class_E'] 
```

```python
#renaming the columns
class_P.columns=['class_type', 'class_P'] 
```

```python
#input: graphing the data from class_E column
#output: bar graph with class_E data
class_E.plot.bar(x = 'class_type', y = 'class_E', title = 'Number of Class E Bike Routes')
```

```python
#input: graphing the data from class_P column
#output: bar graph with class_P data
class_P.plot.bar(x = 'class_type', y = 'class_P', title = 'Number of Class P Bike Routes')
```
